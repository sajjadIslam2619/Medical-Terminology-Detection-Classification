```python
from termcolor import colored
def print_treatment(word):
    return colored(word,'white','on_red', attrs=['underline'])
def print_test(word):
    return colored(word, 'white','on_magenta')
def print_problem(word):
    return colored(word, 'magenta','on_cyan', attrs=['reverse', 'bold'])
```

```python
for sen, lab in zip(sentences, labels):
    for s, l in zip(sen, lab):
        if l in ['B-treatment', 'I-treatment']:
            print(print_treatment(s), end=' ')
        elif l in ['B-test', 'I-test']:
            print(print_test(s), end=' ')
        elif l in ['B-problem', 'I-problem']:
            print(print_problem(s), end=' ')
        else:
            print(s, end=' ')
```
